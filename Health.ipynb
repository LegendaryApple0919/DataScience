```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
def loading_data(file):
    return pd.read_csv(file)

df = loading_data("./Health_heart_experimental.csv")
df.head(10)
```

```python
age = [34,64,39,50,68,36,30,23,21,35]
SysBp = [144,141,140,125,115,113,112,111,109,106]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(SysBp,age, color='green', marker='+', linestyle='dotted')

# add a title
plt.title("Blood Pressure According to Age")

# add a label to the y-axisYears
plt.ylabel("Age")
plt.xlabel("BP")
plt.show()
```

LineGraph According to the varying number of age and their blood pressure. The y plane represents age while the x plane represents their respective BP.

```python
sns.barplot(x = 'sex',
            y = "SysBP",
            data = df)

plt.show()
```

This BarGraph is according to the comparison of Blood Pressure and sex of the people whose data is taken. Both sexes seem to have similar blood pressure. 

```python
sns.barplot(x = 'sex',
            y = "indication",
            data = df)

plt.show()
```

This BarGraph represents Sex in x axis VS indication in y axis. Both sexes seem to have equal number of people with health indications.

```python
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

This is a PieChart representing the sexes of people whose data is taken. Here, it can be observed that although the data of "1" is less by few decimal points, "1" and "0" have nearly equal amount of data.

```python
sns.boxplot(df['age'])
```

```python
sns.boxplot(x='sex',y='age',hue='indication',data=df)
```

```python
sns.kdeplot(df[df['indication'] == 0]['age'], label='indicated')
sns.kdeplot(df[df['indication'] == 1]['age'], label='not indicated')

plt.title('Age Distribution by Indication')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
